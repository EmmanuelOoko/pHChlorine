```python
import matplotlib as plt
from matplotlib import image as img
import cv2 as cv
import pandas as pd
```

# **Free Chlorine 1.0 Manipulation**

```python
fcl_1b = pd.read_csv("/content/chlorine_data/fcl_1.0_rgb.csv")
fcl_1v = pd.read_csv("/content/chlorine_data/fcl_1.0_hsv.csv")
fcl_1b
```

```python
fcl_1b.drop(['Unnamed: 0', 'chlorine'],  axis=1, inplace=True)
fcl_1b.drop(fcl_1b.index[1200:], inplace=True)
fcl_1b.describe()
```

```python
fcl_1v.drop(['Unnamed: 0'],  axis=1, inplace=True)
fcl_1v.rename(columns={'lumination':'value'}, inplace=True)
fcl_1v.drop(fcl_1v.index[1200:], inplace=True)
fcl_1v.describe()
fcl_1v
```

```python
fcl1 = pd.DataFrame()
fcl1 = pd.concat([fcl_1b, fcl_1v], axis=1)
fcl1.to_csv('/content/chlorine_data/fcl1.csv')
fcl1
```
